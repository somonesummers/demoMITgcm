```python
import numpy as np
import gsw
from scipy import interpolate
import matplotlib.pyplot as plt
grid_params = {}
grid_params['Nr'] = 50
grid_params['Ny'] = 12
grid_params['Nx'] = 80


dz = np.ones(grid_params['Nr']) * 20

z = -np.cumsum(dz)

writeFiles = False
```

```python


t2 = np.zeros([grid_params['Nr'],grid_params['Ny'],grid_params['Nx']])
s2 = np.zeros([grid_params['Nr'],grid_params['Ny'],grid_params['Nx']])
S2 = np.zeros([grid_params['Nr'],grid_params['Ny']])
T2 = np.zeros([grid_params['Nr'],grid_params['Ny']])


# z = -[0 100 350 400 500];
# t1 = [0.2 0.2 1.7 1.5 1.1];
# t(:,1) = interp1(z,t1,zprof,'cubic');

# z = -[0 100 200 300 400 500];
# s1 = [32 33.8 34.2 34.3 34.4 34.5];

z_tmp =  np.asarray([  0,  100,  200,  300,  400,   500, 1000]); #must be increasing, so do depth as positive, see negs later for z[:]
t_tmp =  np.asarray([0.2,  0.2,  0.5,  1.7,  1.5,   1.3, 1.0]);
s_tmp =  np.asarray([ 32, 33.8, 34.2, 34.3, 34.4, 34.45,34.5]);
print(z_tmp)
# t_cs = interpolate.CubicSpline(z_tmp, t_tmp)
# s_cs = interpolate.CubicSpline(z_tmp, s_tmp)
t_cs = interpolate.PchipInterpolator(z_tmp, t_tmp)
s_cs = interpolate.PchipInterpolator(z_tmp, s_tmp)
for j in np.arange(0,grid_params['Ny']):
    for i in np.arange(0, grid_params['Nx']):
        t2[:, j, i] = t_cs(-1 * z[:])
        s2[:, j, i] = s_cs(-1 * z[:])

for j in np.arange(0,grid_params['Ny']):
    T2[:,j] = t_cs(-1 * z[:])
    S2[:,j] = s_cs(-1 * z[:])

# write_bin("T2.init", t2)
# write_bin("S2.init", s2)
# write_bin("EBCs2.bin", S2)
# write_bin("EBCt2.bin", T2)

plt.figure()
plt.plot(S2 - 34, z, 'b', label="Sref - 34")
plt.plot(T2, z, 'r', label="Tref")
plt.scatter(s_tmp - 34,-z_tmp,color='b')
plt.scatter(t_tmp,-z_tmp,color='r')
plt.legend()
if(writeFiles):
    plt.savefig("%sinitialTS2" % (run_config['run_dir']+'/input/'))
plt.show()
plt.close()
```

```python
conservT = gsw.conversions.CT_from_t(s2[:,0,0]/1000,t2[:,0,0],-z)
potDensity = gsw.density.sigma0(s2[:,0,0]/1000,conservT)

plt.plot(conservT,z[:],label='conserv T')
plt.plot(t2[:,0,0],z[:],label='In situ T')
plt.plot(t2[:,0,0]-conservT,z[:],label='Difference')
plt.legend()
plt.grid(color='xkcd:light gray')
```
